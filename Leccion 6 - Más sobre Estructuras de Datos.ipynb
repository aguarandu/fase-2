## Funciones de una Lista

| Método |Sintaxis| Descripción   | 
|------|------|-----|
|  clear()   | lista.clear() |Elimina todos los elementos de la lista | 
|  copy()  |lista.copy() |Devuelve una copia de la lista  | 
|  count("elemento")| lita.count("elemento")| Devuelve una cuenta de la cantidad de elementos especificados  | 
|  extend() | lista.extend(otra_lista)| Agrega cada elemento de la otra_lista al final de la lista. La otra_lista puede ser cualquier iterable | 
|index("elemento")|lista.index("elemento")|Devuelve el indice del elemento especificado|
|remove("elemento")|lista.remove("elemento")|Elimina el elemento especificado|
|reverse()|lista.reverse()|Invierte el orden de la lista|
|sort()|lista.sort()|Arregla la lista los elementos de la lista en orden ascendente|



````
ensalada = ["manzana", "uva","banana" ,"pera", "mandarina", "sandia", "banana", "kiwi","uva"]

# copiamos la lista ensalada dentro de la lista nueva_ensalada
nueva_ensalada = ensalada.copy()

# eliminamos todos los elementos de la lista ensalada
ensalada.clear()

# ahora imprimimos nuestro resultado
print("ensalada: ", ensalada,"nueva ensalada: ", nueva_ensalada)

# contamos cuantas veces se repite "banana en nuestra lista"
print("banana se repite: ", nueva_ensalada.count("banana"))

# en que indice se encuentra "mandarina"
print("mandarina se encuentra en el indice: ", nueva_ensalada.index("mandarina"))

# extendemos la lista nueva_ensalada con la lista verduras
print("Lista original: ", nueva_ensalada)
nueva_ensalada.extend(["papaya", "melon", "frutilla"])
print("Lista extendida:", nueva_ensalada)

# la diferencia entre extend y append es que con append agregariamos la lista
# completa, y con extend agregamos los elementos uno a uno
nueva_ensalada.append(["banana", "sandia"])
print("Lista usando append: ", nueva_ensalada)

# tambien podemos agregar un elmento en el indice que queramos
nueva_ensalada.insert(1, "mandarina")
print("Lista usando insert: ", nueva_ensalada)

# podemos sacar un elemento por su nombre
nueva_ensalada.remove(["banana", "sandia"])
print("Lista despues de remove: ", nueva_ensalada)

# ordenamo nuestra lista
nueva_ensalada.sort()
print("Lista ordenada: ", nueva_ensalada)

# invertimos el orden de la lista
nueva_ensalada.reverse()
print("Lista invertida: ", nueva_ensalada)
````

#### copy()
¿Por qué utilizar el método `copy()`  si vimos que en Python podemos asignar un valor a una variable simplemente haciendo: 
```
saludo = "Hola Mundo"
saludo2 = saludo
```
Pues porque cuando hacemos esto:


````
lista_1 = [1,2,3]
lista_2 = lista_1 
print("lista 1: ", lista_1)
print("lista 2: ", lista_2)
````

Parece funcionar no? Pero lo que en realidad está sucediendo es que en la `lista_2` ahora existe la misma `lista_1` solamente que con un nombre diferente o **alias**. Por lo tanto si cambiamos algo dentro de la `lista_1`, este cambio se verá reflejado dentro de la `lista_2`:

````
lista_1 = [1,2,3]
lista_2 = lista_1 
print("lista 1: ", lista_1)
print("lista 2: ", lista_2)
# Ahora sacamos el ultimo elemento de la lista_1: 
print("Vamos a eliminar el ultimo elemento de la lista 1:")
lista_1.pop()

# y volvemos a imprimir ambas listas:
print("lista 1: ", lista_1)
print("lista 2: ", lista_2)
````

Ahora si utilizamos el metodo `copy()`, lo que hacemos es crear una copia de la `lista_1` y esa copia será almacenada en la `lista_2`, por lo tanto la `lista_2`ya no será un **alias** de la `lista_1`: 

````
lista_1 = [1,2,3]
lista_2 = lista_1.copy()
print("lista 1: ", lista_1)
print("lista 2: ", lista_2)

# Ahora sacamos el ultimo elemento de la lista_1: 
print("Vamos a eliminar el ultimo elemento de la lista 1:")
lista_1.pop()

# y volvemos a imprimir ambas listas:
print("lista 1: ", lista_1)
print("lista 2: ", lista_2)
````

Y como puedes ver, esta vez la `lista_2`quedó intacta cuando hicimos el cambio dentro de la `lista_1`

### El Operador `in`

Podemos preguntar si existe un elemento dentro de un arreglo y obtener un valor True o False como respuesta, esto es útil en las condicionales, por ejemplo:


````
nombres = [ "Alex", "Darmy", "Balu", "Ivan"]

if "Darmy" in nombres:
  print("Hola Darmy!")
  
# Otros ejemplos:
colores = ["rojo","naranja", "amarillo"]
print("verde" in colores) # devuelve False porque no existe en la lista
print("rojo" in colores) # devuelve True porque si existe

print(5 in [1,2,3,4]) # 5 no esta en la lista de numeros
print(2 in [1,2,3,4]) # True porque 2 existe en la lista

# El operador in funciona tambien con listas dentro de listas
# en este caso devuelve True 
print(["qwerty"] in [["a","b"], [1,2],["qwerty"]]) 
````

## Listas Anidadas
Las listas pueden contener cualquier tipo de dato, incluso otras listas, a esto le llamamos *Listas Anidadas*:

````
mi_lista = ["cafe", [9,8,3], ["a"], 86421, 3.14]

for elem in mi_lista:
  # usamos este metodo para preguntar si el elemento en la lista
  # es del tipo string
  if isinstance(elem, str):
    print(elem, "es un string")
  
  # usamos este metodo para preguntar si el elemento en la lista
  # es otra lista
  elif isinstance(elem , list):
    print(elem, "es una lista")
    
  elif isinstance(elem, int):
    print(elem, "es un numero decimal o integer")
    
  elif isinstance(elem, float):
    print(elem, "es un numero con coma o float")
    
# str, list, int y float son palabras reservadas de Python, utilizadas
# para hablar del tipo de dato de una variable. Que sea reservada 
# significa que nosotros no podemos utilizarlas para nombres de variables
# o funciones!
````

Como podemos ver, es posible tener listas dentro de listas y además diferentes tipos de datos en una misma lista, el método `isinstance()` junto con las palabras reservadas para cada tipo de dato (string, list, int, float, etc) nos ayudan a evaluar cada tipo de dato y realizar cierta acción al respecto. En el siguiente ejemplo utilizaremos esto para imprimir todos los elementos de las listas anidadas: 

````
mi_lista = ["cafe", [9,8,3], ["a"], 86421, 3.14, ["x","y",["z"]]]

for i in mi_lista:
  # por cada elemento preguntamos si es del tipo lista
  # si es True la recorremos e imprimimos sus elementos
  if isinstance(i, list):
    # todo lo que se imprima despues de esto estara en la misma linea
    print(i,"-> es una lista, sus elementos son:", end=" ")
    
    for elem in i:
      print(elem , end =" ")
    print() # para imprimir un salro de linea
  
  # sino, hacemos un simple print
  else:
    print(i, "-> no es una lista")
````

Analiza el código de arriba, usamos tres conceptos conocidos: bucles anidados, condicionales y la función `isinstance()`.
Con el primer for recorremos sobre cada elemento de la lista y preguntamos si ese elemento es una lista. Si lo es, imprimimos "Los elementos de la lista son:" y comienza el segundo for, donde recorremos los elementos de la lista anidada. Entramos al else cuando en el primer for encontramos elementos que no sean listas!

Pequeño recordatorio, cuando utilizamos `end= " "`  dentro de un `print()` quiere decir que todo se va a imprimir en una misma linea. Cuando hacemos un `print()` vacío, no se imprime nada pero ocurre un salto de línea.

#### Arreglos de dos dimenciones
Cuando tenemos una lista con listas anidadas de una manera uniforme, le llamamos una lista de dos dimensiones. Estas listas anidadas pueden formar una tabla o matriz:

````
coordenadas= [[1,2,3], [4,5,6], [7,8,9]]

# todos los elementos de la lista coordenadas son sublistas de numeros
print("Lista de coordenadas: ", coordenadas)

# primer elemento de la lista coordenadas:
i = 0 
print("elementos en coordenadas")

while i < len(coordenadas):
  print("coordenadas[",i,"]=", coordenadas[i])
  i+=1
  
# al imprimir la lista de coordenadas nos podemos dar cuenta
# de como se asemeja a una tabla donde el primer elemento en 
# la fila 1 y columna 1 es 1 y el útlimo elemento en la fila 3
# y columna 3 es 9
````

Como podemos ver, el primer lemento en la lita de coordenadas es la lista que contine `[ 1, 2, 3 ]`:
```
coordenadas[0]= [1, 2, 3]
```
Sabemos que para acceder a los elementos de una lista utilizamos sus índices, podemos hacer la misma operación la subista `[ 1, 2, 3 ]`. En el siguiente ejemplo utilizaremos el método copy para copiar esta sublista en una nueva variable:


````
coordenadas= [[1,2,3], [4,5,6], [7,8,9]]
print("coordenadas[0] = ",coordenadas[0])

# copiamos el contenido de coordenadas[0] utilizando el metodo copy()
sub_lista = coordenadas[0].copy()
print("sub_lista = ",sub_lista)

# ahora vamos a imprimir el primer elemento en nuestra sublista
print("sub_lista[0]= ", sub_lista[0])
````

Entonces si 

```
coordenadas[0] = [1, 2, 3]
```
y 
```
sub_lista[0] = 1
```
Para acceder al primer elemento de la primera lista anidada debemos especificar dos índices, el primero indica que sublista queremos ver, y el segundo qué elemento de esa sublita queremos obtener: 
```
coordenadas[0][0] = 1
```
De igual manera si queremos el segundo elemento de la sublista hacemos:
```
coordenadas[0][1] = 2
```
En el siguiente programa podemos ver como recorrer todos los elementos de cada sublita utilizando un bucle anidado:

````
coordenadas= [[1,2,3], [4,5,6], [7,8,9]]

# todos los elementos de la lista coordenadas son sublistas de numeros
print("Lista de coordenadas: ", coordenadas)


i = 0 # contador para el bucle exterior
print("Los elementos anidados en la lista de coordenadas son:")

# el bucle se ejecuta hasta llegar al ultimo elemento de la lista
while i < len(coordenadas):
  j = 0 # contador para el bucle anidado
  
  # el bucle se ejecuta hasta llegar al ultimo elemento de la sub-lista
  while j< len(coordenadas[i]):
    print("coordenadas[",i,"][",j,"]=",coordenadas[i][j]) 
    j+=1
  
  i+=1
  
# como dijimos el primer elemento en la lista de coordenadas es una lista, 
# por lo tanto cada elemento de esta lista puede ser accedido por su
# propio indice:
#     coordenadas[0] --> [1,2,3]
#     coordenadas[0][0] --> 1
````

#### *Pista*
Como estas creando una tabla de dos dimensiones, tendrás que rellenar primero la lista anidada con tres elementos, y cuando termines de hacer esto recién agregar la lista anidada a tu tabla, esto lo puedes lograr facilmente con un bucle anidado!  Además cuando hagas tu bucle anidado para rellenar tu tabla ten en mente dos factores:

1.   La diferencia entre utilizar el método `copy()` y asignar directamente el valor de una lista
2.   Que al terminar de rellenar tu lita anidada, sus elementos van a seguir allí, así que deberías vaciarla para volver a rellenarla para evitar tener elementos repetidos.



````
tabla = []
subtabla = []

for i in range(0,3):
  for j in range(0,3):
    movida = input("ingresa tu movida: ")
    subtabla.append(movida)
  
  tabla.append(subtabla.copy())
  subtabla.clear()

print("La tabla de tateti es: ")

print(tabla)
for fila in tabla:
  for elemento in fila:
    print(elemento, end=" ")
  print()
````
