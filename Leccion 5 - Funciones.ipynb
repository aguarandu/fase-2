## Sintaxis


```
def nombre_de_la_funcion ( argumentos, separados, por, coma ):
      """
      Documentación: explicacion para el usuario sobre que hace nuestro metodo (opcional)
      """
      
       sentencias
       return devolver_algo_desde_la_funcion (opcional)
```

La reservada `def` es siempre utilizada al inicio de la definición de una función. Otra nueve palabra reservada es `return`, que nos sirve para *retornar* valores que existen dentro de nuestra función para que puedan ser utilizadas fuera de ella.

Las sentencias dentro de la función no son ejecutadas hasta que *llamamos* a la función, esto se hace escribiendo el nombre de la función, así como lo hacemos siempre con la función `print()`:

En la primera lección de esta fase habíamos hablado de los cinco pasos para programar, dijimos que el último paso era la documentación. Utilizando tres veces la comilla  ( " " " ) comenzamos el bloque de documentación. Cuando otro programador quiera utilizar nuestro código, podrá ver lo que está escrito dentro de este bloque y así entender bien nel código de nuestra función.


````
# Definimos nuestra funcion
def saludo():
  print("Hola desde una función!")
# Fin de la definicion 


# Primero se ejecuta el print y luego la llamada a la funcion "saludo()"
print("Este es un ejemplo de una función muy simple: ")
saludo()
````

Vamos a modificar nuestra función "saludo()" para saludarnos por nuestro nombre, agregando el parámetro nombre de la siguiente manera:

````
# Definimos nuestra funcion
def saludo(nombre):
  print("Hola", nombre, "!")
# Fin de la definicion 

n = input("Ingresa tu nombre: ")
print("Este es un ejemplo de una función muy simple con un solo parametro: ")
saludo(n)
````

Vamos a modificar nuestra función saludo, para retornar un valor!

````
# Definimos nuestra funcion
def saludo(nombre):
  return "Hola "+nombre+"!"
# Fin de la definicion 


n = input("Ingresa tu nombre: ")


# lo que se retorne de la funcion (en este caso un string)
# puede ser guardado dentro de una variable para ser utilizado
# mas tade

s = saludo(n)
print(s)


# pero también puede ser utilizado directamente:
# en este caso el string retornado por nuestra funcion
# fue directamente utilizado por el print

print(saludo(n))
````

El valor que se retorna desde una función puede ser utilizado como una variable cualquiera, en este caso lo utilizamos dentro de una condicional:

````
def suma(a, b):
  resultado = a+b
  return resultado

x = int(input("Ingresa un numero: "))
y = int(input("Ingresa un numero: "))

if suma(x,y) >= 20:
  print("La suma es mayor o igual a 20")
else:
  print("La suma es menor a 20")
````

## Módulos
Un módulo es como un programa de Python que contiene funciones y definiciones útiles. En otras palabras, un módulo se trata de un archivo con código de Python, optimizado para resolver problemas de manera eficiente y ahorrarnos el trabajo de escribir funciones que necesitamos todos los días. Para el siguiente programa utilizaremos el módulo `random` (que significa aleatorio en inglés). 

Para utilizar un módulo debemos traerlo a nuestro programa o en otras palabras *importarlo*, al hacer esto, todas las funciones del módulo estarán disponibles para nosotros. 

Utilizamos la palabra reservada `import` (importar en inglés) seguida del nombre del módulo que queremos utilizar, en este caso es el módulo `random` :
```
import random
```
Listo! luego de hacer esto ya tendremos todas las funciones de este módulo a nuestra disposición, pero la función que nos interesa a nosotros es `randint(a,b)` que genera un número aleatorio entre a y b. Para utilizar esta función debemos escribiendo el nombre del módulo punto el nombre de la función con sus correspondientes parámetros:
```
secreto = random.randint(1, 10)
```
Esto genera un número aleatorio entre el 1 y 10, cada vez que ejecutemos nuestro programa obtendremos un número diferente.

## Juego de adivinanza:


1.   Va a generar un número aleatorio entre el 1 y 10
2.   Va a pedir al usuario que adivine el número
3.   Si el usuario se equivoca el programa debería darle una pista como "el número es muy grande o muy pequeño"
4.   Si el usuario adivinó el número o si ingresa una letra como "q" termina el programa
No te preocupes si esto suena un poco confuso, vamos a crear nuestro programa de a poco y explicando paso a paso lo que sucede.

````
import random

def adivinanza():
  print("Adivina el numero, tienes 3 intentos!\n")
  
  # secreto => numero que debe ser adivinado
  secreto = random.randint(1, 10)
  
  # adivinado => numero que el usuario ingresa
  # intento => cantidad de intentos para adivinar el numero
  adivinado, intentos = 0, 0

  # nuestro bucle while va a solicitar al usuario que ingrese un numero 
  # como hay solo 3 intentos para adivinar, cuando nuestra cuenta de
  # intentos sea 3, termina el programa
  while intentos != 3:
    
    # almacenamos el numero que ingresa el usuario
    adivinado = int(input())
    
    # aumentamos nuestra cuenta de intentos
    intentos += 1
        
    # empezamos el juego de adivinanza:
    if adivinado == secreto:
      print("Hurra adivinaste el numero secreto!")
      
      # un return vacio hace que termine la funcion
      return 

    elif adivinado < secreto:
      print("Este numero es muy chico")
        
    else:
      print("Este numero es muy grande")
        
  # si luego de 3 intentos no se adivino el numero
  # entro al else de nuestro while
  # aca revelo cual era el numero secreto
  else:
    print("\nEl numero secreto era: ", secreto)

adivinanza()
````

Ahora vamos a hacer nuestro programa un poco más complejo, ahora podemos hacer que le juego termine ingresando la letra "q".  Para ello vamos a usar una nueva variable llamada `rpta`, esta va a recibir el input del usuario. Vamos a necesitar un nuevo bloque `if` donde vamos a preguntar si el usuario intresó la letra "q", si lo hizo terminamos el juego, sino empezamos el juego de adivinanza:

````
import random

def adivinanza():
  print("Adivina el numero, tienes 3 intentos! o q para salir del juego\n")
  
  # secreto => numero que debe ser adivinado
  secreto = random.randint(1, 10)
  
  # adivinado => numero que el usuario ingresa
  # intento => cantidad de intentos para adivinar el numero
  adivinado, intentos = 0, 0

  # nuestro bucle while va a solicitar al usuario que ingrese un numero 
  # como hay solo 3 intentos para adivinar, cuando nuestra cuenta de
  # intentos sea 3, termina el programa
  while intentos != 3:
    
    # almacenamos la respuesta del usuario. lower() es un metodo
    # de string para convertir en minusculas la palabra
    rpta = input().lower()
    
    if rpta == "q":
      print("Adios!")
      return
    
    else:
      # si la respuesta no fue q, suponemos que es un numero
      adivinado = int(rpta)
      
      intentos += 1
       
      # empezamos el juego de adivinanza:
      if adivinado == secreto:
        print("Hurra adivinaste el numero secreto!")
        return 

      elif adivinado < secreto:
        print("Este numero es muy chico")
        
      else:
        print("Este numero es muy grande")
        
  # si luego de 3 intentos no se adivino el numero
  # entro al else de nuestro while
  # aca revelo cual era el numero secreto
  else:
    print("\nEl numero secreto era: ", secreto)

adivinanza()
````

¿Qué pasaría si el usuario ingresa algo que no sea un número o la letra "q"? tendríamos un error, ya que el primer if solamente pregunta si se ingresó "q". Ingresa cualquier otra cosa y nuestro programa terminará con un error. Para evitar esto, agregaremos un `elif`. Además antes de iniciar el bloque de adivinanza vamos a verificar que el número ingresado esté entre 1 y 10.

````
import random

def adivinanza():
  """
  Genera aleatoriamente un numero y pide al usuario que lo adivine
  tiene tres oportunidades para hacerlo, si se equivoca el programa 
  le da una pista como "Este numero es muy pequeño". Si el usuario 
  adivina el numero antes de tres intentos, se imprie "Hurra" y termina 
  el programa, sino se imprime el numero secreto. El programa tiene 
  la opcion de ingresar "q" para terminar
  """
  
  print("Adivina el numero, tienes 3 intentos! o q para terminar\n")
    
  secreto = random.randint(1, 10)
  adivinado, intentos = 0, 0

  while intentos != 3:
    rpta = input().lower()
    
    if rpta == "q":
      print("Adios!")
      return
    
    # si no es "q" pero tampoco es un numero, es una respuesta invalida
    # el usuario pierde un intento y continua el bucle
    elif not rpta.isdigit():
      print("Respuesta invalida, perdiste una oportunidad!\n")
      intentos += 1
      continue
    
    # entra aqui si se ingreso un numero  
    else:
      adivinado = int(rpta)
      
      # verificamos que sea un numero del 1 al 10
      if adivinado in range(1,10):
        intentos += 1
        
        # empezamos el juego de adivinanza:
        if adivinado == secreto:
          print("Hurra adivinaste el numero secreto!")
          return 

        elif adivinado < secreto:
          print("Este numero es muy chico")
        
        else:
          print("Este numero es muy grande")
      
      # si no esta en el rango de 1 al 10, es una respuesta invalida
      else:
        print("Respuesta invalida, el numero debe estar entre 1 y 10")
        print("Perdiste una oportunidad!\n")
        intentos += 1
        continue
        
  # si luego de 3 intentos no se adivino el numero
  # entro al else de nuestro while
  # aca revelo cual era el numero secreto
  else:
    print("\nEl numero secreto era: ", secreto)

adivinanza()
````

Cuando el usuario hace un movimiento inválido pierde un intento, para eso escribimos `intentos+=1`

#### Calculadora
Crea una función `calculadora(x, y, op)` que va a recibir como parámentros dos números y qué operación realizar: suma, resta, multiplicación o división. Controla que las operaciones ingresadas sean válidas. Utiliza esta función en un bucle `while` y en cada repetición del bucle le dices al usuario que ingrese `0`si quiere calcular algo, o `1`para terminar el juego.

````
def calculadora(x, y, op):

    if op not in '+-/*':
        return 'Operacion invalida, las operaciones soportadas son: "+, -, *, /"!'

    if op == '+':
      return(str(x) + ' ' + op + ' ' + str(y) + ' = ' + str(x + y))
    if op == '-':
      return(str(x) + ' ' + op + ' ' + str(y) + ' = ' + str(x - y))
    if op == '*':
      return(str(x) + ' ' + op + ' ' + str(y) + ' = ' + str(x * y))
    if op == '/':
      return(str(x) + ' ' + op + ' ' + str(y) + ' = ' + str(x / y))


while True:
  e = int(input("Ingresa 0 para calcular algo, 1 para terminar el programa: "))
  
  if e == 0:
    x = int(input("x: "))
    y = int(input("y: "))
    op = input("operacion: ")
    print(calculadora(x,y,op))
  
  elif e == 1:
    print("Adios!")
    break
````
